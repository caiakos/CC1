#INTRODUÇÃO À LINGUAGEM DE PROGRAMAÇÃO

## **Linguagens de Máquina**

O que distingue os computadores de outras máquinas é que os **computadores podem ser programados**. Isso significa que **Instruções podem ser armazenadas em um arquivo no disco rígido e depois carregadas para a memória principal e executadas por demanda**. Como as máquinas não podem processar a ambiguidade da forma como nós (humanos) podemos, as instruções devem ser precisas. Os computadores fazem exatamente o que lhes é solicitado, e não podem compreender o que o programador “pretendeu” escrever.

As instruções realmente executadas são instruções em **[Linguagem de Máquina](https://en.wikipedia.org/wiki/Machine_code)**. Elas são representadas usando a **Notação Binária (ou seja, uma sequência de 0s e 1s)**.

## **Linguagens de Programação**

Visto que as instruções em linguagem de máquina são extremamente difíceis de se trabalhar, os cientistas da computação desenvolveram **Linguagens de Programação** e **Tradutores de Linguagem**.

<font color="blue">**Ou seja; Linguagens de Programação é um método padronizado para comunicar Instruções (Códigos) para um computador.**</font>

Com isso, os desenvolvedores podem escrever instruções em uma linguagem legível aos humanos e depois fazer com que sejam traduzidas (automaticamente ou não) para a linguagem de máquina.

## **Programa de Computador**

É a formalização de um **Algoritmo** em qualquer Linguagem de Programação, capaz de ser transformada em instruções (de máquinas) que serão executadas por um computador.

## **Tradutores**

Esses **Tradutores de Linguagem de Programação** são conhecidos como **montadores** (ou assemblers), **Compiladores** ou **Interpretadores**, dependendo da linguagem de programação.

### **Compilador**

Classicamente, um [**Compilador**](https://en.wikipedia.org/wiki/Compiler) traduz um programa de uma linguagem textual facilmente entendida por um ser humano para uma linguagem de máquina , específica para um processador e sistema operacional.

image.png

As características são:

VANTAGENS:

*   Os códigos do programa já são traduzidos em código de máquina e, portanto, o **tempo para executar o código é muito menor**;
*   Os arquivos EXE são executados mais rapidamente que os códigos-fonte. Os códigos de objeto (.exe) são salvos e podem ser executados a qualquer momento;
*   Códigos de objetos (file.exe) são mais difíceis para um programador alterar do que o códigos-fonte;
*   O compilador verifica o código-fonte quanto a erros: Léxico e Sintaxe durante a compilação.

DESVANTAGENS:

*   À medida que todo o programa é traduzido, ele usa muito mais memória do computador;
*   É muito mais difícil alterar o programa, por ENGENHARIA REVERSA, sem voltar ao código-fonte ao trabalhar com um compilador;
*   O código objeto precisa ser produzido antes de um arquivo executável final, isso pode ser um processo lento;
*   O código fonte deve estar 100% correto para que o arquivo executável seja produzido.

### **Interpretador**

**[Interpretador](https://en.wikipedia.org/wiki/Interpreter_(computing))** é um programa de computador que executa diretamente instruções escritas em uma linguagem de programação ou script, sem exigir que elas tenham sido previamente compiladas em um programa em linguagem de máquina.

image.png

Um **interpretador** geralmente usa uma das seguintes **estratégias para a execução do programa**:

1. Analise o código-fonte e executa diretamente seu comportamento;
2. Traduza o código-fonte em alguma representação intermediária eficiente ou código-objeto e o executa imediatamente;
2. Execute explicitamente o bytecode pré-compilado armazenado feito por um compilador e combinado com a Máquina Virtual do interpretador.

image.png

As características são:

VANTAGENS:

*   Os intérpretes facilitam muito o trabalho com o código-fonte;
*   Ele traduz uma instrução por vez; portanto, usa uma quantidade mínima de memória;
*   Também é útil no processo de depuração porque o interpretador pode relacionar mensagens de erro à instrução que está sendo executada.

DESVANTAGENS:

*   Cada vez que o programa é executado, o tempo de interpretação é gasto, portanto, um **tempo de execução mais lento**;
*   Os programas interpretados só podem ser executados nos computadores que possuem os respectivos intérpretes.

EXEMPLOS DE LINGUAGEM DE PROGRAMAÇÃO INTERPRETADAS:

## **Linguagem de Alto Nível**

*   Uma linguagem com um nível de abstração relativamente elevado, longe do código de máquina e mais próximo à linguagem humana;
*   O programador de uma linguagem de alto nível não precisa conhecer características do processador ( como Instruções Thread e Registradores Cache Memory Essas características são abstraídas na linguagem de alto nível;
*   De modo que, as linguagens de alto nível não estão diretamente relacionadas à arquitetura do computador.

## **Linguagem de Baixo Nível**

*   Uma linguagem próxima do código de máquina
*   O programador de uma linguagem de baixo nível precisa conhecer características do processador ( como Instruções Thread e Registradores Cache Memory);
*   De modo que, as linguagens de baixo nível estão diretamente relacionadas à arquitetura do computador;
*  **Thread** É um pequeno programa que se autodividi em duas ou mais tarefas É o termo para Encadeamento de Execução

## **Construindo uma LP**

```
import math # ENCAPSULAMENTO + ABSTRAÇÃO = MODULARIZAÇÃO
```

```
def IMC(massa, altura):
  return massa / math.pow(altura, 2);
```

```
# ALGORITMO = ENTRADA + PROCESSAMENTO (ULA) + SAÍDA
# MAIN: ENTRADA
try:
  massa = float(input('Massa (em Quilogramas): '))
  altura = float(input('Altura (em metros).   : '))
  if massa < 0 or altura < 0:
    print('ERRO: de quem?')
  else:
    imc = massa / (altura ** 2)
    print(f'IMC: {imc: .1f}')
    imc = IMC(massa, altura)
    print(f'IMC: {imc: .1f}')
except Exception as ERRO:
  print(f'ERRO: de quem? {ERRO}')
```

*   Por exemplo de **Palavras Reservadas (Key Words)** podemos citar o Conjunto de Instruções Lexemas da Linguagem de Programação Python.

###Analisador Semântico

*   É o processo aplicado pelo Compilador ou pelo Interpretador ao última etapa de transformação do código fonte em Código Objeto;
*   Sendo que, código objeto ou arquivo objeto é o nome dado ao código resultante da Compilação / Interpretação do código fonte Nem todos os Interpretadores criam um código objeto;
*   É a terceira fase da compilação interpretação onde verificam os erros semânticos;
*   Erros que ocorrem em **Tempo de Execução** ([**Runtime**](https://en.wikipedia.org/wiki/Runtime_(program_lifecycle_phase))).

```
1 / 0
```

#### Exemplo de um Erro de Semântica:

```
def IMC(massa, altura):
  return massa / (altura ** 2)
```

```
try:
  Massa = float(input("Massa (em Quilogramas): "))
  Altura = float(input("Altura (em Metros): "))
  # TRATAMENTO DE ERRO:
  if (Massa < 0 or Altura < 0):
    print("Erros nos dados de entrada.\n")
  else:
    # FRONT END + BACK END
    imc = IMC(Massa, Altura)
    print(f"Seu IMC: {imc: .1f}")
except Exception as Erro:
  # REUSO / MODULARIZACAO / ENCAPSULAMENTO (ABSTRACAO)
  print(f'Erro de Exceção: {Erro}')
```

##Paradigmas de Programação

*   É um meio de se classificar as linguagens de programação baseado em suas funcionalidades. As linguagens podem ser classificadas em vários paradigmas;
*   Um paradigma de programação fornece e determina a visão que o programador possui sobre a estruturação e execução do programa;
*   As LP podem ser classificadas em vários paradigmas;
*   <font color="red">**Os Paradigmas implementam o Pensamento Computacional dentro das Linguagens de Programação.**

image.png

**PARADIGMAS IMPERATIVOS:**

*   **PARADIGMA ESTRUTURADO**: é um Paradigma de Programação uma forma de programação de computadores, com ênfase no uso de sub rotinas, laços de repetição, condicionais e estruturas em bloco;
*   **PARADIGMA ORIENTADO A OBJETOS**: é um Paradigma de Programação baseado no conceito de Objetos que podem conter dados na forma de campos, também conhecidos como Atributos e códigos, na forma de procedimentos, também conhecidos como Métodos;
*   **PARADIGMA DE CONSULTA DE BASE DE DADOS**: é um Paradigma de Programação para consulta (Query) e manipulação de Banco de dados.

```
# PARADIGMA OO: PYTHON
class aluno:
  def __init__(self, massa=None, altura=None):
    # ESTADO: ATRIBUTOS
    self.massa = massa
    self.altura = altura

  # COMPORTAMENTO: MÉTODOS
  def calcularIMC(self, massa, altura):
    return massa / (altura ** 2)
```

```
aluno1 = aluno()    # OBJETO
```

```
aluno1.calcularIMC(80.500, 1.80)
```

```
type(aluno1.calcularIMC)
```

```
# MAIN: PROGRAMA PRINCIPAL
try:
  aluno1.massa = float(input("Massa (em Quilogramas): "))
  aluno1.altura = float(input("Altura (em Metros): "))
  # TRATAMENTO DE ERRO:
  if (aluno1.massa <= 0 or aluno1.altura <= 0):
    print("Erros nos dados de entrada.\n")
  else:
    print(f"Seu IMC: {aluno1.calcularIMC(aluno1.massa, aluno1.altura): .1f}")
except Exception as Erro:
  print(f'Erro de Exceção: {Erro}')
```

**PARADIGMAS DECLARATIVOS:**

*   **PARADIGMA FUNCIONAL**: É um paradigma de programação que trata a computação como uma avaliação de funções matemáticas e  que evita estados ou dados mutáveis. Ela enfatiza a aplicação de funções, em contraste da programação imperativa, que enfatiza mudanças no estado do programa. Exemplo de Linguagem: Haskell, [ELM](https://elm-lang.org/);
*   **PARADIGMA LÓGICO**: é um paradigma de programação que faz uso da **lógica matemática**. John McCarthy [1958] foi o primeiro a publicar uma proposta de uso da lógica matemática para programação. Exemplo de Linguagem: [PROLOG](https://swish.swi-prolog.org/example/movies.pl).
